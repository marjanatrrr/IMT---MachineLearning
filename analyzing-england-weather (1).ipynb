# üå¶Ô∏è **England Weather Dataset Overview** üå¶Ô∏è

This dataset provides **comprehensive hourly weather data** for England, covering a decade from **2006 to 2016**. It includes key meteorological parameters, making it ideal for **time-series analysis** and **weather predictions**.

---

### <span style="color: #4CAF50;">üìã **Dataset Columns**:</span>
| **Column Name**          | **Description**                                                      |
|--------------------------|----------------------------------------------------------------------|

```python
# import requierment libarary
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
data= pd.read_csv("/kaggle/input/england/EnglandWeather.csv")
dff=pd.DataFrame(data)
dff
```

```python
# Splitting date and time from "Formatted Date" column and dropping the original column
dff[['date', 'time']] = dff['Formatted Date'].str.split(' ', n=1, expand=True)
dff1 = dff.drop(columns=['Formatted Date'])
dff1
```

```python
print(dff1['date'].unique())
```

```python
dff1.info()
```

```python
# Detect duplicate data in your DataFrame 'dff1'
duplicate_data = dff1[dff1.duplicated()]

# Print the number of duplicated data
print(f'Number of duplicated data: {len(duplicate_data)}')

# Print the duplicated rows
if len(duplicate_data) > 0:
    print("Duplicated data rows:")
    print(duplicate_data)
else:
    print("No duplicated data found.")
```

```python
# Create a copy of the DataFrame 'dff1'
dff2 = dff1.copy()

# Drop duplicates from the copied DataFrame
dff2 = dff1.drop_duplicates()

# Display the DataFrame after removing duplicates
dff2
```

## <span style="color: #FF9800;">‚ö†Ô∏è **Missing Value**</span>

```python
dff2.isna().sum()
```

```python
miss_val = dff2[dff2['Precip Type'].isna()]
miss_val
```

```python
# Counting the types of precipitation in data where temperature is above zero
pre_type_counts = dff2[dff2['Temperature (C)'] >= 0]['Precip Type'].value_counts()
print(pre_type_counts)

# Since the precipitation type for data with temperature above zero is mainly rain, 
# we fill missing values with 'rain'
dff3 = dff2.copy()
dff3.fillna('rain', inplace=True)

# Displaying the final result
dff3
```

```python
# verifing
dff3.isna().sum()
```

```python
# Checking for zero pressure in the dataframe
zero_pressure_count = (dff3['Pressure (millibars)'] == 0).sum()

if zero_pressure_count > 0:
    print(f'Number of records with zero pressure: {zero_pressure_count}')
else:
    print('No records with zero pressure found.')
```

```python
# Removing records where pressure is equal to zero
dff3 = dff3[dff3['Pressure (millibars)'] != 0]

# Displaying the number of remaining records
print(f'Number of remaining records: {len(dff3)}')
```

<span style="color: #2A9DF4;">Records with zero pressure are meaningless, so we remove them to ensure the integrity of our dataset.</span>





```python
dff3
```

```python
# 1. Convert humidity to percentage for better analyze
dff3.loc[:, 'Humidity'] = dff3['Humidity'] * 100
dff3
```

```python
# Create a new 'time' column to keep only hours and minutes
dff3.loc[:, 'time'] = dff3['time'].str.split(':').str[:2].str.join(':')  # Remove seconds

# Display the updated dataframe
print(dff3[['date', 'time']])  # Show only the 'date' and 'time' columns
```

<h2 style="color: #4caf50;">üîç EDA: Exploratory Data Analysis</h2>


```python
correlation = dff3[['Temperature (C)', 'Wind Speed (km/h)', 'Pressure (millibars)', 'Humidity']].corr()

plt.figure(figsize=(5, 5))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, linecolor='black', cbar_kws={"shrink": .8})
plt.title('Correlation Heatmap', fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

### üîç **Correlation Analysis:**

- üå°Ô∏è **Temperature and Humidity:**  
  A **strong negative correlation** of **-0.63**, indicating that as **temperature** increases, **humidity** tends to decrease.

- üå°Ô∏è **Temperature and Pressure:**  
  A **moderate negative correlation** of **-0.31**, suggesting that **temperature** and **pressure** are inversely related to some extent.

- üå¨Ô∏è **Wind Speed:**  
  Shows a **weak correlation** with the other variables, indicating minimal linear relationships.


```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=dff3, x='Humidity', y='Temperature (C)', hue='Precip Type', style='Precip Type', palette='Set1', s=100, alpha=0.7)
plt.title('Temperature vs Humidity', fontsize=16)
plt.xlabel('Humidity (%)', fontsize=14)
plt.ylabel('Temperature (¬∞C)', fontsize=14)
plt.legend(title='Precipitation Type', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

üå°Ô∏è **Temperature Differences for Rain and Snow**:  
- ‚ùÑÔ∏è **Snow**: Typically occurs at lower temperatures (below **0¬∞C**).  
- üåßÔ∏è **Rain**: Observed at temperatures above **0¬∞C**.  

üíß **Humidity and Precipitation Type**:  
- ‚ùÑÔ∏è **Snow**: Mostly occurs in humidity levels ranging from **40% to 100%**.  
- üåßÔ∏è **Rain**: Found across a broader range of humidity levels but is more common at **humidity above 60%**.  

üìä **Overall Relationship Between Temperature and Humidity**:  
- As **humidity increases**, there is a relative **decrease in temperature**.  
- The scatterplot indicates that **precipitation type** significantly influences the distribution of temperature and humidity.  


```python
plt.figure(figsize=(7, 5))
sns.histplot(dff3['Temperature (C)'], kde=True, bins=30, color='green', alpha=0.6)
plt.title('Distribution of Temperature', fontsize=15)
plt.xlabel('Temperature (¬∞C)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Convert 'date' column to datetime format if not already
dff3['date'] = pd.to_datetime(dff3['date'])

# Sort the dataframe by 'date' to ensure correct plotting order
dff3 = dff3.sort_values(by='date')

# Identify numeric columns excluding 'date'
numeric_columns = dff3.select_dtypes(include='number').columns

palette = sns.color_palette("coolwarm", len(numeric_columns))

# Plotting for each numeric column
for i, column in enumerate(numeric_columns):
    plt.figure(figsize=(10, 5))  # Adjusted smaller figure size
    sns.lineplot(data=dff3, x='date', y=column, marker='o', color=palette[i])
    plt.title(f'{column} Over Time', fontsize=16, fontweight='bold')
    plt.xlabel('Date', fontsize=12)
    plt.ylabel(column, fontsize=12)
    plt.xticks(rotation=45, fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
palette = ['#50B8A1', '#FF6F61']
columns = ['Temperature (C)', 'Wind Speed (km/h)', 'Pressure (millibars)', 'Humidity']

# Set the number of rows and columns for the subplots
rows = (len(columns) + 1) // 2  # Number of rows
fig, axes = plt.subplots(rows, 2, figsize=(12, rows * 4))  # Width=12, height proportional to rows

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Plot each column as a violin plot
for i, col in enumerate(columns):
    sns.violinplot(ax=axes[i], data=dff3, x='Precip Type', y=col, palette=palette, inner='quartile')
    axes[i].set_title(f'Violin Plot of {col} by Precipitation Type', fontsize=14)
    axes[i].set_xlabel('Precipitation Type', fontsize=12)
    axes[i].set_ylabel(col, fontsize=12)

# Remove extra axes if the number of columns is odd
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot with transparency, smaller points, and color bar
scatter = ax.scatter(dff3['Temperature (C)'], dff3['Wind Speed (km/h)'], dff3['Pressure (millibars)'], c=dff3['Humidity'], cmap='viridis', s=20, alpha=0.6)

# Color bar for Humidity
cbar = fig.colorbar(scatter, ax=ax, shrink=0.5, aspect=5)
cbar.set_label('Humidity', fontsize=12)

# Adjusting axis labels and title
ax.set_xlabel('Temperature (¬∞C)', fontsize=14, labelpad=10)
ax.set_ylabel('Wind Speed (km/h)', fontsize=14, labelpad=10)
ax.set_zlabel('Pressure (millibars)', fontsize=14, labelpad=10)
plt.title('3D Scatter Plot of Temperature, Wind Speed, and Pressure', fontsize=16, pad=20)

# Adjusting the view
ax.view_init(elev=30, azim=60)

plt.show()
```
